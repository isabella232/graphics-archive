```python
# -*- coding: utf-8 -*-
%matplotlib inline

import pandas as pd
import numpy as np
import math
import matplotlib as mplstyle
import matplotlib.pyplot as plt
import locale
import arrow
from locale import atof
from dateutil import parser
from datetime import datetime
```

```python
def countyFips(countyCode):
    if (countyCode < 10):
        return '1200' + str(countyCode)
    if (countyCode < 100):
        return '120' + str(countyCode)
    if (countyCode > 100):
        return '12' + str(countyCode)

def removeWord(county):
    return county.replace(' County', '')

def properCase(county):
    if (county == 'DESOTO'):
        return 'DeSoto'
    else:
        return county.title()
```

```python
# import data
df_fips_original = pd.read_csv('data/fips.csv')
df_latest = pd.read_csv('data/outage-reports-091317-9am.csv', header=None)
```

```python
df_latest
```

```python
# duplicate fips dataframe and build full fips codes with state + county num then drop unnecessary columns + reorder
df_fips = df_fips_original
df_fips['fips_code'] = df_fips_original['county_code'].apply(countyFips)
df_fips = df_fips.drop(['state','state_code','county_code','Unnamed: 4'], axis=1)

cols = ['fips_code','county_name']
df_fips = df_fips[cols]
df_fips['county_name'] = df_fips['county_name'].apply(removeWord)
# df_fips
```

```python
df_latestClean = df_latest
list(df_latestClean)
df_latestClean.columns = ['county_name','fpl_total','fpl_out','fpl_out_pct','duke_total','duke_out','duke_pct','gulf_total','gulf_out','gulf_pct','tampa_total','tampa_out','tampa_pct','fpu_total','fpu_out','fpu_pct','coops_total','coops_out','coops_pct','muni_total','muni_out','muni_pct','overall_out','overall_total','overall_with_pct','overall_out_pct'
]
df_latestClean
```

```python
df_latestClean['county_name'] = df_latestClean['county_name'].apply(properCase)
df_latestClean
```

```python
df_dataMerge = df_fips.merge(df_latestClean,on='county_name',how='outer')
df_dataMerge = df_dataMerge.rename(columns={'fips_code':'county_fips'})
df_dataMerge
df_dataMerge.to_csv('data/finalData.csv')
```
